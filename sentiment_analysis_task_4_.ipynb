```python
plt.figure(figsize=(6,4))
sns.countplot(x="Positive", data=data, order=data["Positive"].value_counts().index, palette="Set2")
plt.title("Sentiment Distribution in Social Media Data")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x="Borderlands", hue="Positive", data=data, palette="Set1",
              order=data["Borderlands"].value_counts().head(10).index) # top 10 topics
plt.title("Sentiment by Topic")
plt.xticks(rotation=45)
plt.show()
```

```python
from wordcloud import WordCloud

def plot_wordcloud(sentiment):
    text = " ".join(data[data["Positive"] == sentiment]["im getting on borderlands and i will murder you all ,"].astype(str))
    wc = WordCloud(width=800, height=400, background_color="white").generate(text)
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.title(f"Wordcloud for {sentiment} Tweets")
    plt.show()

for s in ["Positive", "Negative", "Neutral", "Irrelevant"]:
    plot_wordcloud(s)
```
